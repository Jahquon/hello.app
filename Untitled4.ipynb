# Data Preprocessing

```python
rehab3 = rehab2.copy()
```

```python
rehab3['Year'] = rehab2['DateIssue'].dt.year
rehab3['Month'] = rehab2['DateIssue'].dt.month
rehab3['Day'] = rehab2['DateIssue'].dt.day
rehab3['Time'] = rehab2['DateIssue'].dt.time
rehab3.drop('DateIssue', inplace = True, axis = 1)
rehab3.head()
```

```python
drop_cols = ['Object ID','Block','VBN','Lot','PermitNum','Time','Address','Neighborhood','Day','Council_District']
```

```python
X = rehab3.drop(columns = drop_cols, errors = 'ignore')
rehab3.shape, X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

num_features = ['Year','Month']
cat_features = ['ExistingUse','ProposedUse']
geo_features = ['Longitude','Latitude']

num_pipeline = Pipeline(steps = [
    #('imputer', SimpleImputer(strategy = 'median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline(steps = [
   # ('imputer', SimpleImputer(strategy = 'most_frequent'))
('onehot', OneHotEncoder(handle_unknown = 'ignore'))
])

geo_pipeline = Pipeline(steps = [
    ('scaler', StandardScaler())
])
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features),
        ('geo', geo_pipeline, geo_features)
    ],
    remainder='drop'
)
```

```python
num_features
```

```python
X_proc = preprocessor.fit_transform(X)
```

```python
def elbow_silhouette(
    X_proc,
    k_min=2,
    k_max=12,
    random_state=42):
  ks = list(range(k_min, k_max + 1))
  inertias, silhouettes = [], []


  for k in ks:
      km = KMeans(n_clusters=k, random_state=random_state, n_init="auto")
      labels = km.fit_predict(X_proc)
      inertias.append(km.inertia_)
      silhouettes.append(silhouette_score(X_proc, labels, sample_size = 5000, random_state=random_state))

      return ks, inertias, silhouettes

```

```python
ks, inertias, silhouettes = elbow_silhouette(
    X_proc,
    k_min=2,
    k_max=12,
    random_state=42

)

best_k = ks[int(np.argmax(silhouettes))]
print(f"Best k by silhouette: {best_k} (score={max(silhouettes):.4f})")
```
